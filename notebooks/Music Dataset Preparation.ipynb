# Music Data Preparation

```python
import pandas as pd
```

Since LastFM dataset is huge, firstly we clear this data. To do this, the procedure in [data-preprocessing.sh](../data-preprocessing.sh) should be followed. In this procedure, we simply remove the songs appeared less than 10 times and take unique songs. 

Below, songs from LastFM listening history and MSD genre dataset are loaded and merged. 

```python
last_fm_listening_hist = pd.read_csv('../data/lastfm-dataset-1K/lastfm_uniq_songs_ge9_wout_c_clean_wout_quotes.csv', sep=';')
last_fm_listening_hist.shape
```

```python
last_fm_listening_hist.columns = ['artist_name_lastfm', 'tra_id', 'title_lastfm']
last_fm_listening_hist.head
```

```python
msd_genre = pd.read_csv('../data/msd_genre_dataset.csv')
msd_genre.shape
```

```python
msd_genre.head
```

```python
last_fm_with_msd = pd.merge(last_fm_listening_hist, msd_genre,  how='inner', left_on=['artist_name_lastfm', 'title_lastfm'], right_on = ['artist_name','title'])
last_fm_with_msd.shape
```

```python
last_fm_with_msd.head
```

```python
last_fm_with_msd.to_csv("last_fm_songs_with_features.csv", sep=';')
```
